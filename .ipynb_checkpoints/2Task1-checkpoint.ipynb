```python
# Loads data using code from activity recognition
import os

data_path = "activity_recognition_dataset/S1_Dataset"

files = os.listdir(data_path)
print(files)
```

```python
import pandas as pd  # use pandas to load data from CSV files

combined_data = []

for file in files:
    
    if file == "README.txt":
        continue
    
    try:
        # load a single sequence from the file
        seq_dataframe = pd.read_csv(os.path.join(data_path, file), header=None, names=["time", "frontal acc", "vertical acc", "lateral acc", "antenna ID", "RSSI", "phase", "frequency", "label"])
        
        # put the ID of the sequence into the dataframe as an extra column
        seq_dataframe['seqID'] = file  # use filename as ID
                
        combined_data.append(seq_dataframe)  # put the data frame into a list
    except:
        print(f"Could not load file {file}. Skipping.")
```

```python
import numpy as np

# the pandas dataframe stores data in a table with headers
input_cols = ["frontal acc", "vertical acc", "lateral acc", "RSSI", "phase", "frequency"]  # column headers for the input features
output_col = "label"  # column header for the output label

# get the relevant columns from the pandas dataframes and convert to numpy arrays
X_by_seq = []  # store a list of numpy arrays containing the input features for each sequence
y_by_seq = []  # store a list of 1-D numpy arrays containing the target activity labels for each sequence

for seq_table in combined_data:
    X_by_seq.append(seq_table[input_cols].values)
    y_by_seq.append(seq_table[output_col].values - 1)    # subtract one from the label so that labels start from 0
    
n_states = np.unique(np.concatenate(y_by_seq)).size  # how many classes/states are there?
```

```python
from sklearn.model_selection import train_test_split

# create train/test split. Sequences are kept complete.
X_by_seq_tr, X_by_seq_test, y_by_seq_tr, y_by_seq_test = train_test_split(X_by_seq, y_by_seq, test_size=0.2, random_state=21)  # You may wish to make further splits of the data or to modify this split.
```

```python
# Record which observations occur given each state (activity label)
X_given_y = [[] for _ in range(n_states)]  # empty list where we will record the observations that occur given each activity label

for s, X_seq in enumerate(X_by_seq_tr):
    for i in range(X_seq.shape[0]):
        state_i = y_by_seq_tr[s][i]
        X_given_y[state_i].append(X_seq[i, :][None, :])
```

```python
from hmmlearn import hmm  # We recommend using this class if builing an HMM

# Record the mean feature values for observations in each state
means = np.zeros((n_states, len(input_cols)))

# Record the variance of feature values for observations in each state
diagonal_covars = np.zeros((n_states, len(input_cols)))

for state in range(n_states):
    means[state] = np.mean(X_given_y[state], axis=0)
    diagonal_covars[state, :] = np.var(X_given_y[state], axis=0)
```

```python
X_tr = np.concatenate(X_by_seq_tr, axis=0)  # combine features into one matrix -- use this as input features for training a classifier
y_tr = np.concatenate(y_by_seq_tr)  # combine target labels into one list -- use this as target labels for training a classifier
```

```python
X_test = np.concatenate(X_by_seq_test, axis=0)  # combine features into one matrix -- use this as input features for testing a classifier
y_test = np.concatenate(y_by_seq_test)  # combine target labels into one list -- use this as target labels for evaluating a classifier
```

```python
# Task 1a - Train Neural Network
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models

model = models.Sequential([
    layers.Dense(32, activation='sigmoid', input_shape=(6,)),
    layers.Dense(64, activation='sigmoid'),
    layers.Dense(4, activation='sigmoid')
])

# Converts labels to be one hot
y_tr_hot = np.array([np.array([1 if i == y else 0 for i in range(4)]) for y in y_tr])
y_test_hot = np.array([np.array([1 if i == y else 0 for i in range(4)]) for y in y_test])

epochs = 50
val_split=0.1
# Compile the model
model.compile(optimizer='adam', 
              loss='categorical_crossentropy', metrics=['acc'])

# Train the model
# Note training appears to be very temperemental. Around 2/3 runs will find nothing,
# but it will learn effectively for the other 1/3 of runs.
history = model.fit(X_tr, y_tr_hot, epochs=epochs, validation_split=val_split)

# Plot training and validation loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Part 1b - tests model performance on training dataset
test_loss, test_accuracy = model.evaluate(X_test, y_test_hot)
print("Test loss", test_loss, "Test accuracy", test_accuracy)
```

```python
# Part 1c - Expirementation with hyperparameters for network performance
def create_model_and_train_model(lr, hidden_layers, activation, batch_size, epochs, val_split, X, y, optimiser):
    # Creates model
    first_layer = [layers.Dense(hidden_layers[0], activation=activation, input_shape=(6, ))]
    hidden_layers_keras = [layers.Dense(neurons, activation=activation) for neurons in hidden_layers[1:]]
    hidden_layers_keras.append(layers.Dense(4, activation='softmax'))
    hidden_layers_keras = list(np.concatenate((first_layer, hidden_layers_keras)))
    mod = models.Sequential(hidden_layers_keras)
    # Compiles model
    mod.compile(optimizer=optimiser(lr), loss='categorical_crossentropy', metrics=['acc'])
    # Trains model
    hist = mod.fit(X, y, epochs=epochs, validation_split=val_split, batch_size=batch_size)
    return (hist, mod)

# Defines hyperparameters to expirement with
lr = [1e-5, 1e-3, 1e-1]
hidden_layers = [[32], [32, 32], [32, 32, 32]]
activation = ['sigmoid', 'relu', 'tanh']
optimisers = [tf.keras.optimizers.Adam, tf.keras.optimizers.SGD, tf.keras.optimizers.RMSprop]
batch_size = [1, 2, 16, 32, 64]
# Trains each model for a set number of epochs for effective comparison
epochs = 5
val_split = 0.1
# Defines a base model to compare these items to
base_lr = 1e-3
base_hidden_layers = [32, 64]
base_activation = 'sigmoid'
base_batch_size = 32
base_optimiser = tf.keras.optimizers.Adam
#base_model_hist, base_model_trained = create_model_and_train_model(lr=base_lr, hidden_layers=base_hidden_layers, 
 #                                              activation=base_activation, batch_size=base_batch_size,
  #                                             epochs=epochs, val_split=val_split, X=X_tr, y=y_tr_hot, 
   #                                       optimiser=base_optimiser)
```

```python
# Tests impact learning rate has on model performance
no_trials = 60
lr_perf_loss = []
for l_rate in lr:
    best_loss = 1000
    best_acc = -1
    best_loss_history = None
    # Performs number of trials to prevent local minima influenicng results too much
    for trial in range(no_trials):
        model_lr_hist, model_lr_trained = create_model_and_train_model(lr=l_rate, 
                                                hidden_layers=base_hidden_layers, 
                                               activation=base_activation, batch_size=base_batch_size,
                                               epochs=epochs, val_split=val_split, X=X_tr, y=y_tr_hot, 
                                          optimiser=base_optimiser)
        loss, acc = model_lr_trained.evaluate(X_test, y_test_hot)
        if loss <= best_loss:
            best_loss = loss
            best_loss_history = model_lr_hist
    lr_perf_loss.append(best_loss_history)
```

```python
# Plots learning rate peformance
for idx in range(len(lr)):
    plt.plot(lr_perf_loss[idx].history['loss'], label=str(lr[idx]))
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title("Impact of Learning Rate on Model Training")
plt.xticks(np.arange(5), ["1", "2", "3", "4", "5"])
plt.legend()
plt.show()
```

```python
# Tests impact number of hidden layers has on model performance
no_trials = 20
mod_perf_loss = []
for mod_arch in hidden_layers:
    best_loss = 1000
    best_acc = -1
    best_loss_history = None
    # Performs number of trials to prevent local minima influenicng results too much
    for trial in range(no_trials):
        model_lr_hist, model_lr_trained = create_model_and_train_model(lr=base_lr, 
                                                hidden_layers=mod_arch, 
                                               activation=base_activation, batch_size=base_batch_size,
                                               epochs=epochs, val_split=val_split, X=X_tr, y=y_tr_hot, 
                                          optimiser=base_optimiser)
        loss, acc = model_lr_trained.evaluate(X_test, y_test_hot)
        if loss <= best_loss:
            best_loss = loss
            best_loss_history = model_lr_hist
    mod_perf_loss.append(best_loss_history)
```

```python
# Plots model peformance based on number of hidden layers
for idx in range(len(hidden_layers)):
    plt.plot(mod_perf_loss[idx].history['loss'], label=str(len(hidden_layers[idx])) + " hidden layers")
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title("Impact of Model Architecture on Model Training")
plt.xticks(np.arange(5), ["1", "2", "3", "4", "5"])
plt.legend()
plt.show()
```

```python
# Tests impact activation function has on model performance
no_trials = 40
act_perf_loss = []
for act in activation:
    best_loss = 1000
    best_acc = -1
    best_loss_history = None
    # Performs number of trials to prevent local minima influenicng results too much
    for trial in range(no_trials):
        model_lr_hist, model_lr_trained = create_model_and_train_model(lr=base_lr, 
                                                hidden_layers=base_hidden_layers, 
                                               activation=act, batch_size=base_batch_size,
                                               epochs=epochs, val_split=val_split, X=X_tr, y=y_tr_hot, 
                                          optimiser=base_optimiser)
        loss, acc = model_lr_trained.evaluate(X_test, y_test_hot)
        if loss <= best_loss:
            best_loss = loss
            best_loss_history = model_lr_hist
    act_perf_loss.append(best_loss_history)
```

```python
# Plots performance based on choice of activation function
for idx in range(len(activation)):
    plt.plot(act_perf_loss[idx].history['loss'], label=activation[idx])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title("Impact of Activation Function on Model Training")
plt.xticks(np.arange(5), ["1", "2", "3", "4", "5"])
plt.legend()
plt.show()
```

```python
# Tests impact optimiser has on model performance
no_trials = 20
opt_perf_loss = []
for opt in optimisers:
    best_loss = 1000
    best_acc = -1
    best_loss_history = None
    # Performs number of trials to prevent local minima influenicng results too much
    for trial in range(no_trials):
        model_lr_hist, model_lr_trained = create_model_and_train_model(lr=base_lr, 
                                                hidden_layers=base_hidden_layers, 
                                               activation=base_activation, batch_size=base_batch_size,
                                               epochs=epochs, val_split=val_split, X=X_tr, y=y_tr_hot, 
                                          optimiser=opt)
        loss, acc = model_lr_trained.evaluate(X_test, y_test_hot)
        if loss <= best_loss:
            best_loss = loss
            best_loss_history = model_lr_hist
    opt_perf_loss.append(best_loss_history)
```

```python
# Plots performance based on choice of optimiser
opt_labels = ['Adam', 'SGD', 'RMSprop']
for idx in range(len(optimisers)):
    plt.plot(act_perf_loss[idx].history['loss'], label=opt_labels[idx])
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title("Impact of Optimiser on Model Training")
plt.xticks(np.arange(5), ["1", "2", "3", "4", "5"])
plt.legend()
plt.show()
```

```python
# Tests impact batch size has on model performance
no_trials = 20
bs_perf_loss = []
for b_s in batch_size:
    best_loss = 1000
    best_acc = -1
    best_loss_history = None
    # Performs number of trials to prevent local minima influenicng results too much
    for trial in range(no_trials):
        model_lr_hist, model_lr_trained = create_model_and_train_model(lr=base_lr, 
                                                hidden_layers=base_hidden_layers, 
                                               activation=base_activation, batch_size=b_s,
                                               epochs=epochs, val_split=val_split, X=X_tr, y=y_tr_hot, 
                                          optimiser=base_optimiser)
        loss, acc = model_lr_trained.evaluate(X_test, y_test_hot)
        if loss <= best_loss:
            best_loss = loss
            best_loss_history = model_lr_hist
    bs_perf_loss.append(best_loss_history)
```
